# Plotly and Cufflinks

```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
from plotly import __version__
import cufflinks as cf
```

```python
# import js libraries
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
# connect js to notebook
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
#create Data
df = pd.DataFrame(np.random.randn(100,4),
                  columns='A B C D'.split())

df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50,]})
df2
```

## Line Plot


```python
df.iplot()
#     Click on legend to draw focus on specific element eg A & C
```

## Scatter Plot

```python
df.iplot(kind='scatter', x='A',y='B',mode='markers',size=20)
```

## Bar Plot

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
# Aggregate function on the df bar with count
df.count().iplot(kind='bar')
```

## Box Plot

```python
df.iplot(kind='box')
```

### 3D Surface plot

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],
                    'y':[10,24,32,20,9],
                    'z':[5,4,3,2,1]})
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

## Histogram

```python
df['A'].iplot(kind='hist',bins=30)
```

### Overlapping Histograms 

Turn column on / off by clicking on the legend

```python
df.iplot(kind='hist',bins=30)
```

## Spread type

```python
df[['A','B']].iplot(kind='spread')
```

## Bubble Plot

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```

## Scatter Matrix

```python
df.scatter_matrix()
```
