# Matrix Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load datasets

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
# iris = sns.load_dataset('iris')
tips.head()
```

```python
flights.head()
```

```python
# for heat maps, indexing / correaltions needs to be established

tc = tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=1)
```

# Cluster maps

```python
sns.clustermap(fp,cmap='coolwarm',
               standard_scale=1,linecolor='white',
               linewidths=1)
```

# Regression Plots

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,col='day',
          aspect=0.6, size=8)

# lmplot > linear model plot
```

# Grids

```python
# import iris dataset
# iris = sns.load_dataset('iris')

```

```python
g = sns.PairGrid(tips)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Facet Grid

```python
fg = sns.FacetGrid(data=tips,col='time', row='smoker')

fg.map(sns.distplot, 'total_bill')
```
