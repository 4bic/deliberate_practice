# K - Means clustering basics.

```python
import seaborn as sns
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
from sklearn.datasets import make_blobs
# make blobs allows for creation of artificial data to experiment on
```

```python
data = make_blobs(n_samples=200, n_features=2, 
                  centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1], 
            c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(data[0])
```

```python
km.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K means (K=3)')
ax1.scatter(data[0][:,0],data[0][:,1], c=km.labels_,cmap='rainbow')

ax2.set_title('Originals')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1],cmap='rainbow')
```
